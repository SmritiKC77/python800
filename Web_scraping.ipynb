```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.worldometers.info/gdp/nepal-gdp/"
```

```python
html = requests.get(url).content
```

```python
print(html)
```

```python
# use html. parser to convert it to html dom using BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
# now we want to search tbody from dom or html.
```

```python
table_body = soup.find('tbody') # singlr data lai find multiple data ko lagi findall
```

```python
table_rows = table_body.find_all('tr')
```

```python
table_rows[0]
```

```python
table_row= table_rows[0]
[item.text for item in table_row.find_all('td') ]
```

```python
data = [[item.text for item in row.find_all('td')] for row in table_rows]
data
```

```python
for row in table_rows:
  for tr in row.find_all('td'):
    print(tr.text)
  break
```

```python
column_name = [item.text for item in soup.find_all('th')]
column_name
```

```python
import pandas as pd
df = pd.DataFrame(data, columns = column_name)
df
```

```python
df.head()# first five roes
```

```python
df.tail() # last five roes
```

```python
df.shape # no of rows, no of columns
```

```python
df.to_csv('gdp-nepal.csv', index = False)
```
